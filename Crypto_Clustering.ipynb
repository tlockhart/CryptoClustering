```python
# Import required libraries and dependencies
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import hvplot.pandas  # Ensure hvPlot is imported
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
# Generate summary statistics
display(market_data_df.describe())
display(market_data_df.columns)
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

market_scaled = StandardScaler().fit_transform(market_data_df[market_data_df.columns])
       
crypto_names = display(market_data_df.index)
```

```python
# Create a DataFrame with the scaled data
#IMPORTANT: Only transforms the columns, not the index, so create a new col for the index
market_scaled_df = pd.DataFrame(market_scaled, columns=market_data_df.columns)

# Copy the crypto names from the original data
# Interpretation: Add the coin_id to the market_scaled_df
market_scaled_df['coin_id'] = market_data_df.index

# Set the coinid column as index (must store returned value back)
market_scaled_df = market_scaled_df.set_index('coin_id')

# Display sample data
display(market_scaled_df.head())
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for i in k:
    # 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i ,n_init='auto', random_state=0) 

    # 2. Fit the model to the data using the scaled DataFrame (market_scaled_df)
    k_model.fit(market_scaled_df)
    
    # 3. Append the k_model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)

# [Optional: Calculate the percentage decreases and print them:]
for i in range(1, len(k)):
    percentage_decrease = (inertia[i-1] - inertia[i]) / inertia[i-1] * 100
    print(f"Percentage decrease from k={k[i-1]} to k={k[i]}: {percentage_decrease:.2f}%")

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": k,
    "inertia": inertia
}


# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.plot.line(x="k",
                   y="inertia",
                   title="Elbow Curve",
                   xticks=k)
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:** To determine the best  k  value, we can examine the percent decrease in inertia for each  k  value. The optimal  k  is typically where the percent decrease begins to slow significantly, as indicated by smaller gaps between consecutive percent decreases. Additionally, we can confirm the best  k  value by identifying the elbow point on the Elbow Curve plot, where the rate of decrease in inertia noticeably slows.

In this case, there is a significant drop in inertia from  k = 1  (287.00) to  k = 2  (198.57), and again from  k = 2  to  k = 3  (145.63). After  k = 4  (79.02), the rate of decrease in inertia slows down considerably, indicating diminishing returns in clustering quality. Therefore, the best  k  value is **4**, as suggested by the elbow curve.

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
crypto_model = KMeans(n_clusters=4, n_init='auto', random_state=1)
```

```python
# Fit the K-Means model using the scaled data
crypto_model.fit(market_scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = crypto_model.predict(market_scaled_df)

# View the resulting array of cluster values.
k_4
```

```python
# Create a copy of the DataFrame
crypto_transformed_predictions_df = market_scaled_df.copy()

# Display the DataFrame
display(crypto_transformed_predictions_df.head())
```

```python
# Add a new column to the DataFrame with the predicted clusters
crypto_transformed_predictions_df["crypto_cluster"] = k_4
# display(market_data_df.head())

crypto_transformed_predictions_df["coin_id"] = market_data_df.index

# Display sample data
crypto_transformed_predictions_df.set_index("coin_id", inplace=True)
crypto_transformed_predictions_df.head()
```

```python
# Create a scatter plot using Pandas plot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Use "rainbow" for the color to better visualize the data.
crypto_transformed_predictions_df.plot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="crypto_cluster",
    colormap="rainbow")
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled (market_scaled_df) DataFrame to reduce to three principal components.
crypto_pca = pca.fit_transform(market_scaled_df)

# View the first five rows of the DataFrame. 
crypto_pca[:5]
```

```python
# Retrieve the explained variance (ratio) to determine how much information  can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 37% + 35% +18% = 90%

```python
# Create a new DataFrame with the PCA data.
# Note: The code for this step is provided for you (Look in the PCA Examples where we transformed the data[11.3/Class3/02-Ins_PCA)
# Creating a DataFrame with the PCA data
crypto_pca_df = pd.DataFrame(
    crypto_pca,
    columns=["PCA1", "PCA2", "PCA3"]
)

# Copy the crypto names from the original data (Either will work)
crypto_pca_df["coin_id"] = market_scaled_df.index
# crypto_pca_df["coin_id"] = market_data_df.index

# Set the coinid column as index
crypto_pca_df.set_index("coin_id", inplace=True)

# Display sample data
crypto_pca_df.head(10)
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k = list(range(1, 11))

# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using PCA DataFrame.
# 3. Append the k_model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, n_init='auto', random_state=0)
    k_model.fit(crypto_pca_df)
    inertia.append(k_model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow.head(10)
```

```python
# Plot a (/n Elbow Curve) line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.plot.line(x="k",
                   y="inertia",
                   title="Elbow Curve",
                   xticks=k)
```

#### Answer the following questions: 
* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for  k  is typically where the Elbow Point occurs on the Elbow Curve. This is the point at which the reduction in inertia (or within-cluster variance) starts to slow down significantly, forming a bend or elbow. This indicates a balance between the complexity of the model (number of clusters) and the amount of variance explained. In the Elbow Curve above, the best  k  value appears to be **4**, as this is where the curve flattens noticeably, suggesting diminishing returns from adding more clusters.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The  k -values determined from the original K-Means model data and those selected by the PCA-transformed data are the same ***(4)***. This suggests that the PCA process effectively preserved the underlying structure of the data during dimensionality reduction. As a result, the clustering accuracy and interpretability were not compromised by the reduction process.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
crypto_pca_model = KMeans(n_clusters=4, n_init='auto', random_state=0)
```
